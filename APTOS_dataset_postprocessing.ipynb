```python
from PIL import Image
import os

def get_image_sizes(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Filter out only image files (you can add more extensions if needed)
    image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    
    # Dictionary to store image sizes
    image_sizes = {}
    
    # Get size for each image
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        with Image.open(image_path) as img:
            width, height = img.size
            image_sizes[image_file] = (width, height)
    
    return image_sizes

if __name__ == "__main__":
    folder_path = "/home/shashank/All_code/Research/APTOS_DDPM/generatedSample_192_zoomed"
    
    image_sizes = get_image_sizes(folder_path)
    
    for image_file, size in image_sizes.items():
        print(f"Image: {image_file}, Size: {size[0]} x {size[1]} pixels")
```

#### For merging all the generated image in a single folder and giving a unique identifier to each of them.

```python
# For merging all the generated image in a single folder and giving a unique identifier to each of them.

import os
import shutil
import uuid

def copy_images(source_dir, destination_dir):
    # Create the destination directory if it doesn't exist
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    # Recursively search for image files and copy them
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            # Check if the file is an image (you can extend this list for more image formats)
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
                source_file_path = os.path.join(root, file)
                # Generate a unique identifier for each image
                unique_id = str(uuid.uuid4())[:8]
                # Get the file extension
                file_extension = os.path.splitext(file)[1]
                # Create a new filename with the unique identifier
                new_filename = f"{os.path.splitext(file)[0]}_{unique_id}{file_extension}"
                # Copy the image file to the destination directory with the new filename
                shutil.copy(source_file_path, os.path.join(destination_dir, new_filename))
                print(f"Copied: {source_file_path} to {new_filename}")

# Specify your source directory containing subfolders with images
source_directory = '/home/shashank/All_code/Research/APTOS_DDPM/1_extras/generated_images_370_800timesteps'
# Specify the destination directory where all images will be copied
destination_directory = '/home/shashank/All_code/Research/APTOS_DDPM/all_generated_size_128_classwise/temp370'

# Call the function to copy images
copy_images(source_directory, destination_directory)
```

### Code for creating label inside csv for class wise image

```python
import os
import csv

# Define the folder path containing the images
folder_path = r"D:\VSCode_Workspace\DiabeticRetinopathy\APTOS_128\all_generated_193"
# Initialize a set to store unique id_codes
existing_id_codes = set()

# Check if the CSV file already exists
csv_file_path = r"D:\VSCode_Workspace\DiabeticRetinopathy\APTOS_128\train_128.csv"
if os.path.exists(csv_file_path):
    # If the CSV file exists, read the existing id_codes
    with open(csv_file_path, mode='r') as file:
        reader = csv.reader(file)
        try:
            next(reader)  # Skip the header row
            for row in reader:
                existing_id_codes.add(row[0])  # Add id_code to the set
        except StopIteration:
            pass  # Empty CSV file, no header row to skip

# List to store new image names and labels
new_data = []

# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Adjust file extensions as needed
        id_code = filename
        if id_code not in existing_id_codes:
            # Add the image name and label to the new_data list
            # Assuming the label is an integer
            new_data.append([id_code, 3])  # Replace "1" with your desired integer label
            # Add the id_code to the set of existing id_codes
            existing_id_codes.add(id_code)

# Count the total number of new entries
total_new_entries = len(new_data)

# Append new data to CSV file
with open(csv_file_path, mode='a', newline='') as file:
    writer = csv.writer(file)
    # Write new data
    writer.writerows(new_data)

print(f"{total_new_entries} new entries added to the CSV file.")
```

```python
# Code for couting the number of images in each folder 

import os

def count_images(folder_path):
    # Initialize a counter for the number of images
    image_count = 0
    
    # Iterate through all files in the folder
    for file_name in os.listdir(folder_path):
        # Check if the file is a regular file (not a directory) and if it's an image file
        if os.path.isfile(os.path.join(folder_path, file_name)) and file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
            # Increment the image count
            image_count += 1
    
    return image_count

# Specify the path to the folder containing images
folder_path = r'D:\VSCode_Workspace\DiabeticRetinopathy\APTOS_128_original\Processed_Images_balanced'
# Call the function to count the number of images
num_images = count_images(folder_path)
print(f'Total number of images in the folder: {num_images}')
```

### Resizing Image

```python
import os
from PIL import Image

def resize_images(dataset_folder, output_folder, target_size=(128, 128)):
    # Create the output folder if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)
    
    # List all files in the dataset folder
    files = os.listdir(dataset_folder)
    
    # Iterate over each file
    for file in files:
        # Construct the full path to the image file
        file_path = os.path.join(dataset_folder, file)
        
        # Check if the file is an image (with supported extension)
        if file.endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
            # Open the image
            img = Image.open(file_path)
            
            # Resize the image
            resized_img = img.resize(target_size, Image.ANTIALIAS)
            
            # Construct the path for saving the resized image
            output_path = os.path.join(output_folder, file)
            
            # Save the resized image
            resized_img.save(output_path)

# Example usage
dataset_folder = r'D:\VSCode_Workspace\DiabeticRetinopathy\APTOS_128_original\Processed_Images_balanced'
output_folder = r'D:\VSCode_Workspace\DiabeticRetinopathy\APTOS_128\Processed_Images_128'
resize_images(dataset_folder, output_folder)
```
