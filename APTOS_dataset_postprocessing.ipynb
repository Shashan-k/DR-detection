```python
from PIL import Image
import os

def get_image_sizes(folder_path):
    # List all files in the folder
    files = os.listdir(folder_path)
    
    # Filter out only image files (you can add more extensions if needed)
    image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    
    # Dictionary to store image sizes
    image_sizes = {}
    
    # Get size for each image
    for image_file in image_files:
        image_path = os.path.join(folder_path, image_file)
        with Image.open(image_path) as img:
            width, height = img.size
            image_sizes[image_file] = (width, height)
    
    return image_sizes

if __name__ == "__main__":
    folder_path = "/home/shashank/All_code/Research/APTOS_DDPM/generatedSample_192_zoomed"
    
    image_sizes = get_image_sizes(folder_path)
    
    for image_file, size in image_sizes.items():
        print(f"Image: {image_file}, Size: {size[0]} x {size[1]} pixels")
```

```python
#### Zooming and Cropping ###

from PIL import Image
import os

def zoom_in_image(image_path, zoom_percentage):
    # Open the image
    image = Image.open(image_path)
    
    # Calculate new dimensions based on zoom percentage
    width, height = image.size
    new_width = int(width * (1 + zoom_percentage / 100))
    new_height = int(height * (1 + zoom_percentage / 100))
    
    # Resize the image without changing the canvas size
    zoomed_image = image.resize((new_width, new_height))
    
    # Calculate coordinates for cropping to original size
    left = (new_width - width) // 2
    top = (new_height - height) // 2
    right = left + width
    bottom = top + height
    
    # Crop the zoomed image to original size
    zoomed_image = zoomed_image.crop((left, top, right, bottom))
    
    return zoomed_image

def main(input_folder, output_folder, zoom_percentage):
    # Create the output folder if it doesn't exist
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    # List all files in the input folder
    files = os.listdir(input_folder)
    
    # Filter out only image files (you can add more extensions if needed)
    image_files = [f for f in files if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    
    # Zoom in and save the images
    for image_file in image_files:
        input_image_path = os.path.join(input_folder, image_file)
        output_image_path = os.path.join(output_folder, image_file)
        
        zoomed_image = zoom_in_image(input_image_path, zoom_percentage)
        zoomed_image.save(output_image_path)
        
        #print(f"Zoomed in and saved {image_file} to {output_folder}")

if __name__ == "__main__":
    input_folder = "/home/shashank/All_code/Research/APTOS_DDPM/class_4_resized_192"
    output_folder = "/home/shashank/All_code/Research/APTOS_DDPM/class_4_192_zoomed"
    zoom_percentage = 35  # You can change this value to adjust the zoom percentage
    
    main(input_folder, output_folder, zoom_percentage)
```
