```python
import os
import shutil
import pandas as pd
from sklearn.model_selection import train_test_split

# Set the paths
image_folder = '/home/shashank/All_code/Research/APTOS_dataset/aptos2019_blindness_detection/Real_Synthetic/RS_train_all'
csv_file = '/home/shashank/All_code/Research/APTOS_dataset/aptos2019_blindness_detection/Real_Synthetic/RS_train_all.csv'
train_folder = '/home/shashank/All_code/Research/APTOS_dataset/aptos2019_blindness_detection/Real_Synthetic/RS_train_split'
val_folder = '/home/shashank/All_code/Research/APTOS_dataset/aptos2019_blindness_detection/Real_Synthetic/RS_val_split'
train_csv = '/home/shashank/All_code/Research/APTOS_dataset/aptos2019_blindness_detection/Real_Synthetic/RS_train_split.csv'
val_csv = '/home/shashank/All_code/Research/APTOS_dataset/aptos2019_blindness_detection/Real_Synthetic/RS_val_split.csv'

# Load the CSV file
df = pd.read_csv(csv_file)

# Split the data into train and validation sets
train_ids, val_ids = train_test_split(df['id_code'], test_size=0.2, random_state=42)

# Create the train and validation folders if they don't exist
os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)

# Copy the images to the respective folders
for img_file in os.listdir(image_folder):
    for train_id in train_ids:
        if img_file.startswith(train_id):
            shutil.copy(os.path.join(image_folder, img_file), os.path.join(train_folder, img_file))
            break
    for val_id in val_ids:
        if img_file.startswith(val_id):
            shutil.copy(os.path.join(image_folder, img_file), os.path.join(val_folder, img_file))
            break

# Save the train and validation sets as CSV files
train_df = df[df['id_code'].isin(train_ids)]
val_df = df[df['id_code'].isin(val_ids)]

train_df.to_csv(train_csv, index=False)
val_df.to_csv(val_csv, index=False)
```

#### Just Merging mutiple folders into a single folder

```python
# Initialize a counter for the image filenames
import os
import shutil

# Set the source and destination directories
source_dir = "D:/Projects/Research/DDPM/Synthetic_Data_all_classwise"
destination_dir = "D:/Projects/Research/DDPM/Synthetic_Data_All"

# Create the destination folder if it doesn't exist
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# Iterate through the source folders
for folder_name in os.listdir(source_dir):
    folder_path = os.path.join(source_dir, folder_name)
    if os.path.isdir(folder_path):
        # Iterate through the files in the current folder
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            # Copy the file to the destination folder
            shutil.copy2(file_path, os.path.join(destination_dir, filename))
```

### Counting the number of images

```python
# Code for couting the number of images in each folder 

import os

def count_images(folder_path):
    # Initialize a counter for the number of images
    image_count = 0
    
    # Iterate through all files in the folder
    for file_name in os.listdir(folder_path):
        # Check if the file is a regular file (not a directory) and if it's an image file
        if os.path.isfile(os.path.join(folder_path, file_name)) and file_name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif')):
            # Increment the image count
            image_count += 1
    
    return image_count

# Specify the path to the folder containing images
folder_path = '/home/shashank/All_code/Research/APTOS_DDPM/all_generated_size_128_classwise/all_generated_999'
# Call the function to count the number of images
num_images = count_images(folder_path)
print(f'Total number of images in the folder: {num_images}')
```

### Sharpen the Image


```python
#Import the necessary libraries 
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 

# Load the image 
image = cv2.imread(r'D:\Projects\Research\sample_sharpness\image_1_45f69b81.png') 

#Plot the original image 
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

# Create the sharpening kernel 
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) 

# Sharpen the image 
sharpened_image = cv2.filter2D(image, -1, kernel) 

#Save the image 
cv2.imwrite('sharpened_image.jpg', sharpened_image) 

#Plot the sharpened image 
plt.subplot(1, 2, 2) 
plt.title("Sharpening") 
plt.imshow(sharpened_image) 
plt.show()


```
